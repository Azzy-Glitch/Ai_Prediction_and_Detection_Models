```python
import gradio as gr
import os
from transformers import BertForSequenceClassification, BertTokenizer, pipeline

# 1. Setup path
model_path = os.path.abspath("news_topic_model")
print(f"Directly loading from: {model_path}")

# 2. Load model
model = BertForSequenceClassification.from_pretrained(model_path)
tokenizer = BertTokenizer.from_pretrained(model_path)

# 3. Pipeline
classifier = pipeline(
    "text-classification",
    model=model,
    tokenizer=tokenizer
)

# 4. Labels
labels = {
    "LABEL_0": "World News",
    "LABEL_1": "Sports",
    "LABEL_2": "Business",
    "LABEL_3": "Sci/Tech"
}

def predict(text):
    if not text.strip():
        return "⚠️ Please enter some text."
    results = classifier(text)
    top_pred = results[0]
    friendly_name = labels.get(top_pred['label'], top_pred['label'])
    score = round(top_pred['score'] * 100, 2)
    return f"Prediction: {friendly_name} ({score}%)"

# 5. UI (compatible version)
with gr.Blocks() as demo:

    # Header
    gr.Markdown("# 📰 News Topic Classifier (BERT)")
    gr.Markdown("Classify headlines into World, Sports, Business, or Sci/Tech")

    # Input Section
    gr.Markdown("### 📝 Enter Headline")
    input_text = gr.Textbox(
        lines=3,
        placeholder="Type your news headline here..."
    )

    # Buttons
    with gr.Row():
        predict_btn = gr.Button("🔍 Analyze")
        clear_btn = gr.Button("🧹 Clear")

    # Output Section
    gr.Markdown("### 📊 Result")
    output_text = gr.Textbox(interactive=False)

    # Examples
    gr.Markdown("### 💡 Examples")
    gr.Examples(
        examples=[
            ["Stock markets surge after tech rally"],
            ["New AI breakthrough in robotics"],
            ["Football team wins championship"],
            ["Global leaders meet for climate summit"]
        ],
        inputs=input_text
    )

    # Events
    predict_btn.click(fn=predict, inputs=input_text, outputs=output_text)
    input_text.submit(fn=predict, inputs=input_text, outputs=output_text)
    clear_btn.click(lambda: ("", ""), outputs=[input_text, output_text])

# Launch
if __name__ == "__main__":
    demo.launch(share=True)
```
