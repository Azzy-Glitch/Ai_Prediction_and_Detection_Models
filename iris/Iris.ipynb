```python
# ============================================================
# Task 1: Exploring and Visualizing a Simple Dataset (Iris)
# ============================================================
# Objective:
# The objective of this task is to load, inspect, and visualize
# the Iris dataset in order to understand data trends, feature
# distributions, and relationships between variables.
#
# Tools Used:
# - pandas for data handling
# - seaborn and matplotlib for visualization
# ============================================================
```

```python
# -------------------------------
# Step 1: Import Required Libraries
# -------------------------------
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python

# -------------------------------
# Step 2: Load the Iris Dataset
# -------------------------------
# The Iris dataset is loaded directly from seaborn.
# It is stored in a pandas DataFrame.
df = sns.load_dataset("iris")
```

```python
# -------------------------------
# Step 3: View First Few Rows
# -------------------------------
# head() displays the first 5 rows of the dataset
df.head()
```

```python
# -------------------------------
# Step 4: Dataset Shape
# -------------------------------
# shape shows the number of rows and columns
print("DataSet Shapes: ")
for shp in df.shape:
    print(f"- {shp}")
```

```python
# -------------------------------
# Step 5: Column Names
# -------------------------------
print("\nDataset Columns:")
for col in df.columns:
    print(f"- {col}")
```

```python
# -------------------------------
# Step 6: Dataset Information
# -------------------------------
# info() provides details about data types and missing values
df.info()
# Dataset Information
# The dataset contains 150 observations of iris flowers with 4 numerical features
# and 1 categorical feature (species).
# There are no missing values in the dataset, which makes it clean and ready for analysis.
```

```python
# -------------------------------
# Step 7: Summary Statistics
# -------------------------------
# describe() generates statistical summary of numerical features
df.describe()
# Summary Statistics
# The summary statistics show the mean, standard deviation, and range of each feature.
# Petal length and petal width show greater variation compared to sepal measurements,
# which may help in distinguishing different species.
```

```python

# -------------------------------
# Step 8: Data Visualization
# -------------------------------

# 8.1 Scatter Plot
# Shows relationship between sepal length and petal length
sns.scatterplot(x = 'sepal_length', y = 'petal_length', hue = 'species', data = df)
plt.title("Sepal Length vs Petal Length by Species")
plt.show()
# The scatter plot between sepal length and petal length shows a clear separation
# among species.Setosa is distinctly separated from the other two species,
# while Versicolor and Virginica overlap slightly.
```

```python
# 8.2 Histograms
# Shows distribution of numerical features
df.hist(figsize = (10, 8))
plt.suptitle("Feature Distributions")
plt.show()

# The histograms show that most features follow a roughly normal distribution.
# Petal-related features show clearer grouping patterns compared to sepal features.
```

```python
# 8.3 Box Plot
# Used to detect outliers and understand data spread
plt.figure(figsize = (10, 6))
sns.boxplot(data = df)
plt.title("Box Plot for Outlier Detection")
plt.show()
# Boxplots help in identifying outliers and spread of the data.Petal width and
# petal length have noticeable differences across species, confirming their importance.
```

```python
# -------------------------------
# Final Observations
# -------------------------------
# The scatter plot shows a clear separation between species based on petal
# and sepal measurements.
# Histograms show normal distribution for most features.
# Boxplots help identify any potential outliers in the dataset.
# Petal length and petal width are strong indicators for distinguishing species.
```
